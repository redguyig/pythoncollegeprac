### Computer Vision Pipeline
  - It has many steps in it:
     1) Image Acquisition - acquiring the image and converting it to 2d array
     2) Pre Processing of image - resizing to fit the mesh, noise reduction, contrast adjustification 
     3) Segmentation - seperating the main object from bg 
     4) Features Extraction and selection - acquiring image and extracting their main features
     5) High-level Processing - Basically processing additional features like if we extact a face then processing it's features such as hair color, skin color, etc
