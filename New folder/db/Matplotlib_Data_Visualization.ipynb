```python
### Bar graph
student=["A","B","C","D"]
marks=[23,43,49,32]

plt.bar(student,marks,width=0.4)
plt.title("Python result")
plt.xlabel("Students name")
plt.ylabel("Marks")
plt.show()
```

```python
student=["A","B","C","D"]
marks=[23,43,49,32]
exp=[0,0,0.2,0]
plt.pie(marks,labels = student, explode=exp,colors=["red","green","yellow","black"],shadow=True,autopct="%1.1f%%",startangle=90)
plt.title("Python result")
plt.show()
```

```python
### Bar graph
plt.figure(figsize=(10,10))
student=["A","B","C","D"]
marks_python=[23,43,49,32]
#marks_maths=[22,43,33,28]
student1=["E","F","G","H"]
marks_python1=[23,43,49,32]
plt.bar(student,marks_python,width=0.4)
plt.bar(student1,marks_python1,width=0.4)
plt.title("Python result")
plt.xlabel("Students name")
plt.ylabel("Marks")
plt.show()
```
